```python
from flask import Flask, request, send_from_directory
import os, pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
import m2cgen as m2c
def variance(values):
    print(values)
    n= len(values)
    variance=0
    mean=sum(values)/n
    for row in values:
        variance+=(row-mean)**2
    variance=variance/n
    return variance
    
app = Flask(__name__)
MODEL_DIR = "./model"
os.makedirs(MODEL_DIR, exist_ok=True)

@app.route('/upload/edge_three', methods=['POST'])
def upload():

    # 1. Salva CSV ricevuto
    filename = f"edge_three.csv"
    with open(filename, "w") as f:
        f.write(request.data.decode())
    print("Dati ricevuti:\n", filename) 
    # 3. Preprocessing
    df = pd.read_csv(filename)
    print(df.head())
    X=df[['rssi']].values
    model = OneClassSVM(nu=0.5, gamma='scale')
    model.fit(X)
   

    # 4. Esporta in Python puro
    code = m2c.export_to_python(model)
    path = os.path.join(MODEL_DIR, f"model_edge_three.py")
    with open(path, "w") as f:
        f.write(code)

    return f"Trained model for edge_three", 200

@app.route('/model/edge_three')
def serve():
    return send_from_directory(MODEL_DIR, f"model_edge_three.py")


@app.route('/upload/edge_imu', methods=['POST'])
def upload_imu():
    treshold=0.5
    # 1. Salva CSV ricevuto
    filename = f"edge_imu.csv"
    with open(filename, "w") as f:
        f.write(request.data.decode())
    print("Dati ricevuti:\n", filename) 
    # 3. Preprocessing
    df = pd.read_csv(filename,sep=",")
    print(df.head())

   
   
    
    X = df[["AccX","AccY","AccZ","GyroX","GyroY","GyroZ","AngX","AngY","AngZ","MagX","MagY","MagZ"]].values
    model = OneClassSVM()
    model.fit(X)
  
   

    # 4. Esporta in Python puro
    code = m2c.export_to_python(model)
    path = os.path.join(MODEL_DIR, f"model_edge_imu.py")
    with open(path, "w") as f:
        f.write(code)

    return f"Trained model for edge_imu", 200

@app.route('/model/edge_imu')
def serve_imu():
    return send_from_directory(MODEL_DIR, f"model_edge_imu.py")

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=8000)

```
